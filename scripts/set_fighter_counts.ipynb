# Fighter Count Per Set

This notebook loads fighter data and counts how many fighters belong to each set.

```python
from pathlib import Path
import json
from collections import Counter

# Prefer the primary fighters file; fall back if needed
candidate_paths = [Path('input/fighters.json'), Path('input/fighters1.json')]
data_path = next((p for p in candidate_paths if p.exists()), None)

if data_path is None:
    raise FileNotFoundError('No fighters file found in input/.')

with data_path.open('r', encoding='utf-8') as f:
    data = json.load(f)

fighters = data.get('fighters', [])
set_counts = Counter(
    fighter.get('set', 'Unknown Set')
    for fighter in fighters
    if isinstance(fighter, dict)
)

print(f'Using file: {data_path}')
print(f'Total fighters: {len(fighters)}')
print('\nFighters per set:')
for set_name, count in sorted(set_counts.items()):
    print(f'- {set_name}: {count}')
```
